### Formatando os valores da coluna YearMonth

## Análise pix

### Crescimento no número de transações pix desde 2020

```python
qtd_transacoes_pix_2020 = base_dados.loc[base_dados['YearMonth'].dt.year == 2020, 'quantityPix'].sum()
qtd_transacoes_pix_2021 = base_dados.loc[base_dados['YearMonth'].dt.year == 2021, 'quantityPix'].sum()
qtd_transacoes_pix_2022 = base_dados.loc[base_dados['YearMonth'].dt.year == 2022, 'quantityPix'].sum()
qtd_transacoes_pix_2023 = base_dados.loc[base_dados['YearMonth'].dt.year == 2023, 'quantityPix'].sum()
qtd_transacoes_pix_2024 = base_dados.loc[base_dados['YearMonth'].dt.year == 2024, 'quantityPix'].sum()

x = ['2020', '2021', '2022', '2023', '2024']
y = [qtd_transacoes_pix_2020, qtd_transacoes_pix_2021, qtd_transacoes_pix_2022, qtd_transacoes_pix_2023, qtd_transacoes_pix_2024]

# Defini o tamanho do gráfico
plt.figure(figsize=(15,6))

# bars recebe cada barra do gráfico
bars = plt.bar(x, y)

# Adiciona o valor das transações em texto acima de cada barra no gráfico de barras
for bar in bars:
    yval = bar.get_height()  # Obtém a altura da barra atual, que representa o valor das transações
    # Adiciona o texto no centro horizontal da barra e ligeiramente acima da sua altura
    plt.text(bar.get_x() + bar.get_width() / 2.0, yval, f'{yval:.2f}', va='bottom', ha='center')

plt.title("Crescimento no número de transações pix")
plt.xlabel("Anos")
plt.ylabel("Quantidade de transações (escala em bilhões)")
plt.show()
```

### Valores movimentados por pix desde 2020

```python
val_mov_pix_2020 = base_dados.loc[base_dados['YearMonth'].dt.year == 2020, 'valuePix'].sum()
val_mov_pix_2021 = base_dados.loc[base_dados['YearMonth'].dt.year == 2021, 'valuePix'].sum()
val_mov_pix_2022 = base_dados.loc[base_dados['YearMonth'].dt.year == 2022, 'valuePix'].sum()
val_mov_pix_2023 = base_dados.loc[base_dados['YearMonth'].dt.year == 2023, 'valuePix'].sum()
val_mov_pix_2024 = base_dados.loc[base_dados['YearMonth'].dt.year == 2024, 'valuePix'].sum()

x = ['2020', '2021', '2022', '2023', '2024']
y = [val_mov_pix_2020, val_mov_pix_2021, val_mov_pix_2022, val_mov_pix_2023, val_mov_pix_2024]

# Defini o tamanho do gráfico
plt.figure(figsize=(15,6))

# bars recebe cada barra do gráfico
bars = plt.bar(x, y)

# Adiciona o valor das transações em texto acima de cada barra no gráfico de barras
for bar in bars:
    yval = bar.get_height()  # Obtém a altura da barra atual, que representa o valor das transações
    # Adiciona o texto no centro horizontal da barra e ligeiramente acima da sua altura
    plt.text(bar.get_x() + bar.get_width() / 2.0, yval, f'R$ {yval:.2f}', va='bottom', ha='center')

plt.title("Valores movimentados por pix desde 2020")
plt.xlabel("Anos")
plt.ylabel("Valores movimentados (escala em bilhões)")
plt.show()
```

### Variação no volume de transações pix feitas ano passado

```python
ano_passado = datetime.date.today().year -1
x = ['Jan', 'Fev', 'Mar', "Abr", "Mai", "Jun", "Jul", "Ago", "Set", "Out", "Nov", "Dez"]
y = base_dados.loc[base_dados['YearMonth'].dt.year == ano_passado, 'quantityPix'].sort_index(ascending=False)

# Defini o tamanho do gráfico
plt.figure(figsize=(15,6))

# Adiciona a grade atrás do gráfico
plt.grid(True, which='both', linestyle='-', linewidth=0.5, zorder=0)

# Cria o gráfico e defini os estilos
plt.plot(x,y, linestyle='--', marker='o')

plt.title(f"Variação no volume de transações feitas em {ano_passado}")
plt.xlabel("Meses")
plt.ylabel("Número de transações (escala em bilhões)")
plt.show()
```

### Chat

```python
chat = model.start_chat(history=[])

base_dados_IA = base_dados

opcaoSelecionada = ''

while opcaoSelecionada != "2":

  opcaoSelecionada = input("""
  1 - Conversar com o gemini
  2 - Sair
  """)

  if opcaoSelecionada == "1":
    if base_dados_IA.to_string(index=False) != '':
      prompt = input('Converse com o gemini sobre a base de dados: ')
      prompt_completo = f"Aqui estão alguns dados: \n{base_dados_IA.to_string(index=False)}\n\n{prompt}"
      response = chat.send_message(prompt_completo)
      print("\n\nResposta:", response.text, '\n\n')
    else:
      print("\n\nA base de dados está vázia!\n\n")
      opcaoSelecionada == "2"
  elif opcaoSelecionada == "2":
      break
  else:
    print("\n\nOpção Inválida!\n\n")
```
